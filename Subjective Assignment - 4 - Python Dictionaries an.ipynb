Q6. Write the Python program to push three items into the heap and
return the smallest item from the heap. Also, return and pop the
smallest item from the heap?
    
    Expected OutputItems in the heap-
    ('VI', 1).
    ('VII', 3).
    ('VIII', 2).
    ----------------------
    The smallest item in the heap:
    ('VI', 1).
     ----------------------
    Pop the smallest item in the heap:
    ('VII', 2).
    ('VIII', 3).

```python
class BinHeap:
    def __init__(self):
        self.heapList = [0]
        self.currentSize = 0
        
        
    def heapify_downtop(self, index):
        
        while index // 2 > 0:            
            if self.heapList[index] < self.heapList[index // 2]:
                self.heapList[index], self.heapList[index // 2] = self.heapList[index // 2], self.heapList[index]
            index //= 2
    
    def heapify_topdown(self, index):
        
        while (index * 2) < self.currentSize:
            if self.heapList[index] > self.heapList[2 * index]:
                if self.heapList[2 * index] > self.heapList[2 * index + 1]:
                    self.heapList[index], self.heapList[2 * index + 1] = self.heapList[2 * index + 1], self.heapList[index]
                else:
                    self.heapList[index], self.heapList[2 * index] = self.heapList[2 * index], self.heapList[index]
            index *= 2
            
    def insert(self,k):
        
        self.heapList.append(k)
        self.currentSize += 1
        index = self.currentSize
        self.heapify_downtop(index)
        
    def pop(self):
        
        index = self.currentSize
        self.currentSize -= 1
        self.heapList[1] = self.heapList[index]
        self.heapList.pop()
        self.heapify_topdown(1)
        return self.heapList
        
    def findmin(self):
        return self.heapList[1]
    
    
b = BinHeap()
b.insert(('VI', 1))
b.insert(('VII', 3))
b.insert(('VIII', 2))
print("Items in the heap" + "--"*10)
print(b.heapList)
print("The smallest item in the heap:" + "--"*10)
print(b.findmin())
print("Pop the smallest item in the heap:" + "--"*10)
print(b.pop())
```

Q8. Write the Python program to create the FIFO queue?

    Expected Output:
    0 1 2 3

```python
class Queue(object):
    
    def __init__(self):
        self.list1 = []
        
    def enqueue(self, item):
        self.list1.append(item)
        return self.list1
        
    def dequeue(self):
        del self.list1[0]
        return self.list1
    

q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.dequeue()
```

Q9. Write the Python program to calculate the harmonic sum of n-1.
 Note: The harmonic sum is the sum of reciprocals of the positive
Integers?

    Example:
    1+1/2+1/3+1/4+1/5+....

```python
def func(k):
    sum = 0
    for i in range(1,k):
        sum += 1/i
    return sum

func(100)
```
